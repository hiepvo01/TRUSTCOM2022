```python
import torch
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
losses = torch.load('results/losses/loss_25.pt')
# 14500
d_r1 = []
d_r2 = []
d_r3 = []
d_f1 = []
d_f2 = []
d_f3 = []
g_1 = []
g_2 = []
g_3 = []
for i in range(len(losses.keys())):
    d_r1.append(losses[i+1][0])
    d_r2.append(losses[i+1][1])
    d_r3.append(losses[i+1][2])
    d_f1.append(losses[i+1][3])
    d_f2.append(losses[i+1][4])
    d_f3.append(losses[i+1][5])
    g_1.append(losses[i+1][6])
    g_2.append(losses[i+1][7])
    g_3.append(losses[i+1][8])
    
data = {'d_r1': d_r1, 'd_r2': d_r2, 'd_r3': d_r3, 'd_f1': d_f1, 'd_f2': d_f2, 'd_f3': d_f3, 'g_1': g_1, 'g_2': g_2, 'g_3': g_3}
df = pd.DataFrame.from_dict(data)
df = df.groupby(np.arange(len(df.index))//10).mean()
df.head()
```

```python
df_r = df[['d_r1', 'd_r2', 'd_r3']]
df_r.plot.line()
```

```python
df_f = df[['d_f1', 'd_f2', 'd_f3']]
df_f.plot.line()
```

```python
df_g = df[['g_1', 'g_2', 'g_3']]
df_g.plot.line()
```
